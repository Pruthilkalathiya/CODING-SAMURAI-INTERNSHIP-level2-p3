```python
# ===========================================================
# üßæ SALES DATA ANALYSIS PROJECT - CODING SAMURAI INTERNSHIP
# Author: Pruthil Kalathiya
# ===========================================================

# --- Step 1: Import Required Libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: make plots display clearly
plt.style.use('seaborn-v0_8')
sns.set_palette("viridis")

# --- Step 2: Load Dataset ---
# (Upload your CSV file named 'sales_data.csv' before running this cell)
df = pd.read_csv("sales_data.csv")

# Display top 5 rows
print("‚úÖ Dataset Loaded Successfully!")
df.head()

# --- Step 3: Basic Info ---
print("\nüìã Dataset Information:")
print(df.info())
print("\nüîç Missing Values:\n", df.isnull().sum())

# --- Step 4: Add 'Revenue' Column ---
df['Revenue'] = df['Quantity'] * df['Price']
print("\nüí∞ 'Revenue' column added successfully!")
df.head()

# --- Step 5: Descriptive Statistics ---
print("\nüìä Descriptive Statistics:")
print(df.describe())

# --- Step 6: Total Revenue Calculation ---
total_revenue = df['Revenue'].sum()
print("\nüí∏ Total Revenue Generated: ‚Çπ", total_revenue)

# --- Step 7: Top-Selling Products ---
product_sales = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
print("\nüèÜ Top Selling Products:\n", product_sales)

# --- Step 8: Category-wise Sales ---
category_sales = df.groupby('Category')['Revenue'].sum()
print("\nüìÇ Revenue by Category:\n", category_sales)

# --- Step 9: Daily Revenue Trend ---
daily_sales = df.groupby('Date')['Revenue'].sum().reset_index()
print("\nüìÜ Daily Revenue (First 5 days):")
print(daily_sales.head())

# --- Step 10: Visualizations ---
# Bar Chart - Product Revenue
plt.figure(figsize=(10,5))
sns.barplot(x=product_sales.index, y=product_sales.values)
plt.title("üí∞ Total Revenue by Product", fontsize=14)
plt.xlabel("Product")
plt.ylabel("Revenue (‚Çπ)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Pie Chart - Category Revenue
plt.figure(figsize=(5,5))
category_sales.plot.pie(autopct='%1.1f%%', colors=['lightblue', 'lightgreen'])
plt.title("üõí Revenue by Category")
plt.ylabel("")
plt.show()

# Line Chart - Daily Revenue Trend
plt.figure(figsize=(10,5))
sns.lineplot(x='Date', y='Revenue', data=daily_sales, marker='o')
plt.title("üìà Daily Revenue Trend")
plt.xlabel("Date")
plt.ylabel("Revenue (‚Çπ)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Heatmap - Correlation Between Numeric Columns
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("üî• Correlation Heatmap")
plt.show()

# --- Step 11: Monthly Analysis (Optional Advanced Step) ---
df['Month'] = pd.to_datetime(df['Date']).dt.strftime('%B')
monthly_sales = df.groupby('Month')['Revenue'].sum().sort_values()
print("\nüóìÔ∏è Monthly Revenue:\n", monthly_sales)

plt.figure(figsize=(6,4))
sns.barplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title("üìÜ Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue (‚Çπ)")
plt.show()

# --- Step 12: Key Insights ---
print("\nüìà INSIGHTS & CONCLUSIONS")
print("--------------------------------------------------")
print(f"‚Ä¢ Total Revenue Generated: ‚Çπ{total_revenue}")
print(f"‚Ä¢ Best Selling Product: {product_sales.index[0]} (‚Çπ{product_sales.iloc[0]})")
print(f"‚Ä¢ Top Performing Category: {category_sales.idxmax()} (‚Çπ{category_sales.max()})")
print("‚Ä¢ Accessories contribute around {:.1f}% of total revenue.".format(
    (category_sales['Accessories']/total_revenue)*100))
print("‚Ä¢ Daily sales peaked around mid-January 2024.")
print("--------------------------------------------------")

print("\n‚úÖ Project Completed Successfully! üéâ")
```
